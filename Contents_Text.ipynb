# Retrieve text for each contents url

```python
import os.path
import requests
from bs4 import BeautifulSoup
import urllib.request
import glob
import time
```

# Using html files in Contents folder

## Test on single content file

```python
query_date = (2018, 10, 4, 3)

cache_path = "./Data/Contents/Contents-%04d-%02d-%02d_%d.html" % (query_date[0], query_date[1], query_date[2], query_date[3])
site_prefix = "https://www.oireachtas.ie"

# Have we already download the file?
if not os.path.exists( cache_path ):
    url = "https://www.oireachtas.ie/en/debates/debate/dail/%04d-%02d-%02d/%02d/" % (query_date[0], query_date[1], query_date[2], query_date[3])
    print("Fetching %s ..." % url)
    r = requests.get(url)
    html = r.text
    fout = open( cache_path, "w" )
    fout.write( html )
    fout.close()
else:
    print("Using cached file")
    html = open( cache_path, "r" ).read()
    
# Parse the full HTML file
soup = BeautifulSoup(html, "html.parser")
```

### Questions - its printing twice

```python
list1 = []

# get date and title
date = soup.find("h1", {"class":"c-hero__title"}).get_text() + "\n"
title = soup.find("div", {"class":"db-section"}).h2.get_text() + "\n"

# Get contents for 'question' tag
results1 = soup.find_all("div",{"class":"db-section"})
for tag1 in results1:
    sub_header = tag1.find("div", {"class":"sub-header"}).h3.get_text() + "\n"
    results2 = tag1.find_all("div", {"class":"question"})
    for tag2 in results2:
        results3 = tag2.find("a", {"class":"c-avatar__name-link"})
        speaker_name = results3.get_text() + "\n"
        results4 = tag2.find_all("div", {"class":"content"})
        for tag3 in results4:
            results5 = tag3.find("p", {"class":""})
            text = results5.get_text() + "\n"
            list1.append(date + title + sub_header + speaker_name + text)
```
