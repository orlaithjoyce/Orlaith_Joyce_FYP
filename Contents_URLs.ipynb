# urls for contents for each date

```python
import os.path
import requests
from bs4 import BeautifulSoup
import urllib.request
import glob
```

### Get URLs for questions for each date

### below is printing some of them twice

```python
query_date = (2018, 10, 4)

titles = ""
links = ""

cache_path = "./Data/Debates/Debates-%04d-%02d-%02d.html" % ( query_date[0], query_date[1], query_date[2] )
site_prefix = "https://www.oireachtas.ie"

# Have we already download the file?
if not os.path.exists( cache_path ):
    url = "https://www.oireachtas.ie/en/debates/debate/dail/%04d-%02d-%02d/" % ( query_date[0], query_date[1], query_date[2] )
    print("Fetching %s ..." % url)
    r = requests.get(url)
    html = r.text
    fout = open( cache_path, "w" )
    fout.write( html )
    fout.close()
else:
    print("Using cached file")
    html = open( cache_path, "r" ).read()


# Parse the full HTML file
soup = BeautifulSoup(html, 'html.parser')

# Find the nested pattern div -> li -> a
results = soup.find_all("div", {"class":"results"})
for tag1 in results:
    results2 = tag1.find_all("li")
    for tag2 in results2:
        results3 = tag2.find_all("a")
        for tag3 in results3:
            link = tag3.get("href","")
            # check the link
            if link.startswith("/"):
                # get the text describing the link
                title = tag3.text.strip()
                titles += title +"\n"
                print(title)
                
                # create the full link
                link = site_prefix + link
                links += link + "\n"
                print(link)
                
                titles_out = open("Data/ContentsTitles", "w" )
                titles_out.write(titles) 
                
                links_out = open("Data/ContentsLinks", "w" )
                links_out.write( links ) 
```

### Save each url to separate files

```python
content_links = open("./Data/ContentsLinks")
n=0
for i in content_links:
    lines = requests.get(i)
    html_data = lines.text
    #use the end of the url (from date onwards as the filename)
    #out_data = open("./Data/Contents/Contents-{}.html".format(i.split("/")[7:]), "w")
    out_data = open("./Data/Contents/Contents-{}.html".format(i[49:59]+"-"+str(n)), "w")
    out_data.write(html_data)
    n+=1
```
